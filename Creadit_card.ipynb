```python
#d1.drop(['Use Chip'],axis=1,inplace=True)
#d1.drop(['Card_Category'],axis=1,inplace=True)
d1.drop(['Client_Num'],axis=1,inplace=True)
d1.drop(['Week_Start_Date'],axis=1,inplace=True)
d1.drop(['Week_Num'],axis=1,inplace=True)
d1.drop(['Qtr'],axis=1,inplace=True)
```

```python
X=d1.iloc[:,:-1]
```

```python
y=d1.iloc[:,-1]
```

```python
X.head()
```

```python
y.head()
```

```python
#dividing data into train test split
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.33,random_state=42)
```

```python
X_train
```

```python
y_train
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logistic_reg=LogisticRegression()
```

```python
logistic_reg.fit(X_train,y_train)
```

```python
#prediction
```

```python
y_pred = logistic_reg.predict(X_test)
```

```python
#evaluation
from sklearn.metrics import accuracy_score
```

```python
accuracy=accuracy_score(y_test,y_pred)
```

```python
accuracy
```

```python
#confusion matrices
```

```python
from sklearn.metrics import confusion_matrix
```

```python
con_matrx=confusion_matrix(y_test,y_pred)
```

```python
con_matrx
```

```python
# report
```

```python
from sklearn.metrics import classification_report
```

```python
class_report=classification_report(y_test,y_pred)
print(class_report)
```
